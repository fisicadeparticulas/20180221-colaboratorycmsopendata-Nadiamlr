```


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

%matplotlib inline
```

```
muons = pd.read_csv('http://opendata.cern.ch/record/545/files/Dimuon_DoubleMu.csv')
muons.head()
```

```
len(muons)
```

```
fig = plt.figure(figsize=(15, 10))

plt.hist(muons.M, 300, range = (0,150))

plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 15)
plt.ylabel('Number of events \n', fontsize = 15)
plt.title('Measured distribution of events with two muons \n', fontsize = 15)

plt.show()
```

```
bump = muons[(muons.M >= 80) & (muons.M <= 100)]

fig = plt.figure(figsize=(15, 10))

plt.hist(bump.M, 300, range = (80,100))

plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 15)
plt.ylabel('Number of events \n', fontsize = 15)
plt.title('Measured distribution of events with two muons \n', fontsize = 15)

plt.show()
```

```
sample7 = pd.read_csv('https://goo.gl/PU9w7m')
sample7.head()
```

```
invariant_mass = np.sqrt(2*sample7.pt1*sample7.pt2*(np.cosh(sample7.eta1-sample7.eta2) - np.cos(sample7.phi1-sample7.phi2)))
```

```
print(invariant_mass)
```

```
fig = plt.figure(figsize=(15, 10))

plt.hist(invariant_mass, 300, range = (0,150))

plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title('Invariant mass for Z â†’ e + e âˆ’ \n', fontsize = 20)

plt.show()
```

```
delPhi =np.abs(sample7.phi1-sample7.phi2)
```

```
print(delPhi)
```

```
fig = plt.figure(figsize=(15, 10))

plt.hist(delPhi, 300, range = (0,8),color='blue')

plt.xlabel('Delta Phi', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title('Delta Phi', fontsize = 20)

plt.show()
```

```
mass = np.sqrt(2*sample7.pt1*sample7.pt2*(np.cosh(sample7.eta1-sample7.eta2) - np.cos(delPhi)))
s=[]

for i in range(0,len(mass)):
  if delPhi[i]<math.pi:
    s.append(mass[i])
    
fig = plt.figure(figsize=(15, 10))

plt.hist(s, 300, range = (0,150), color='mediumseagreen')

plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title('Measured distribution of events with two electrons \n', fontsize = 20)

plt.show()
```

```
isoT=sample7.isoTrack1+sample7.isoTrack2
```

```
fig = plt.figure(figsize=(15, 10))

plt.hist(isoT, 300, range = (0,150), color='orange')

plt.xlabel('IsoTrack', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title('Isolation in the Tracker \n', fontsize = 20)

plt.show()
```

```
Pt=sample7.pt1+sample7.pt2
```

```
rel_isoT=isoT/Pt
```

```
fig = plt.figure(figsize=(15, 10))

plt.hist(rel_isoT, 300, range = (0,1.3), color='cadetblue')

plt.xlabel('Relative Isolation in the Tracker', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title('Relative Isolation in the Tracker \n', fontsize = 20)

plt.show()
```

```
isoE=sample7.isoEcal1+sample7.isoEcal2
```

```
fig = plt.figure(figsize=(15, 10))

plt.hist(isoE, 300, range = (0,100), color='hotpink')

plt.xlabel('Isolation in the ECAL', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title('Isolation in the ECAL \n', fontsize = 20)

plt.show()
```

```
rel_isoE=isoE/Pt
```

```
fig = plt.figure(figsize=(15, 10))

plt.hist(rel_isoE, 300, range = (0,2), color='pink')

plt.xlabel('Relative Isolation in the ECAL', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title('Relative Isolation in the ECAL \n', fontsize = 20)

plt.show()
```

```
isoH=sample7.isoHcal1+sample7.isoHcal2
```

```
fig = plt.figure(figsize=(15, 10))

plt.hist(isoH, 300, range = (0,100), color='purple')

plt.xlabel('Isolation in the HCAL', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title('Isolation in the HCAL \n', fontsize = 20)

plt.show()
```

```
rel_isoH=isoH/Pt
```

```
fig = plt.figure(figsize=(15, 10))

plt.hist(rel_isoH, 300, range = (0,1), color='hotpink')

plt.xlabel('Relative Isolation in the HCAL', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title(' Relative Isolation in the HCAL \n', fontsize = 20)

plt.show()
```

```
invmass = np.sqrt(2*sample7.pt1*sample7.pt2*(np.cosh(sample7.eta1-sample7.eta2) - np.cos(delPhi)))
l=[]

for i in range(0,len(mass)):
  if delPhi[i]<math.pi and rel_isoT[i]<0.2 and rel_isoE[i]<0.5 and rel_isoH[i]<0.2 and np.abs(sample7.eta1[i])<2.4 and np.abs(sample7.eta2[i])<2.4 and sample7.sigmaEtaEta1[i]<0.02 and sample7.sigmaEtaEta2[i]<0.02:
    l.append(invmass[i])
    
fig = plt.figure(figsize=(15, 10))

plt.hist(l, 300, range = (0,150), color='mediumseagreen')

plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 20)
plt.ylabel('Number of events \n', fontsize = 20)
plt.title('Measured distribution of events with two electrons \n', fontsize = 20)

plt.show()
```
